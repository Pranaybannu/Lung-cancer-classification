```python
import tensorflow as tf
import numpy as np
from PIL import Image
import google.generativeai as genai
```

```python
model= tf.keras.models.load_model("3class.keras")
 
```

```python
##image preprocessing
def preprocess_image(image_path, target_size=(256,256)):
    image = Image.open(image_path).convert("RGB")
    print("original image mode:", image.mode)
    image= image.resize(target_size)
    image = np.array(image)
    print("Shape after conversion to array:", image.shape)
    if len(image.shape) == 2:
        image = np.stack((image,)*3, axis=-1)
        image= image / 255.0
    image = np.expand_dims(image,axis=0)
    print("final processed image shape;", image.shape)
    return image
image_path= "test/normal/7.png"
input_image= preprocess_image(image_path)
```

```python
##PREDICT CANCER TYPE
prediction= model.predict(input_image)
predicted_class= np.argmax(prediction)
cancer_class= {0: "BENINGN", 1: "MALIGNANT", 2: "NORMAL"}
cancer_type= cancer_class.get(predicted_class, "unknown")
print(f"predicted cancer type: {cancer_type}")
```
